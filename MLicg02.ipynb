### Comparison of models

Poređenje rezultata modela sa rezultatima modela iz naučnog rada [Neural Image Caption Generation with Visual Attention](https://www.researchgate.net/publication/272194766_Show_Attend_and_Tell_Neural_Image_Caption_Generation_with_Visual_Attention).

```python
import matplotlib.pyplot as plt
models = ['Google NIC', 'Log Bilinear', 'Soft-Attention', 'Hard-Attention', 'My Model']
```

### Poređenje BLEU-1

```python
bleu_scores1 = [63, 65.6, 67, 67, 51.55]

# Plot BLEU scores
plt.figure(figsize=(10, 6))
plt.bar(models, bleu_scores1, color=['blue', 'yellow', 'red', 'green', 'black'])
plt.ylabel('BLEU-1 Score')
plt.title('BLEU-1 Score Comparison')
plt.ylim(0, 100)
for i, v in enumerate(bleu_scores1):
    plt.text(i, v + 1, str(v), ha='center', va='bottom', fontsize=10)
plt.show()

```

### Poređenje BLEU-2

```python
bleu_scores2 = [41, 42.4, 44.8, 45.7, 28.81]

# Plot BLEU scores
plt.figure(figsize=(10, 6))
plt.bar(models, bleu_scores2, color=['blue', 'yellow', 'red', 'green', 'black'])
plt.ylabel('BLEU-2 Score')
plt.title('BLEU-2 Score Comparison')
plt.ylim(0, 100)  # Setting Y-axis limit for better visualization
for i, v in enumerate(bleu_scores2):
    plt.text(i, v + 1, str(v), ha='center', va='bottom', fontsize=10)
plt.show()
```

### Razlika u modelima

Glavna razlika između ova dva modela je korišćenje "Attend" mehanizma. Dok "Show and Tell" modeli generišu opis na osnovu fiksne reprezentacije cele slike, "Show, Attend and Tell" modeli se fokusiraju na različite delove slike dok generišu svaku reč u opisu, što omogućava potencijalno tačnije i kontekstualno relevantnije opise.

Razlika u modelima je manje primetna kod "jednostanijih" slika, dok su performanse kod složenijih slika dosta bolje za modele koji koriste ovah mehanizam, jer mogu da procenjuju relevantnost različitih regiona slike tokom generisanja opisa.
